```python
#Mejora, se crea bloques de 1.000 asi que en lugar de 100.000 ciclos, hacemos 100 ciclos de 1,000 filas (107 segundos)
import numpy as np
filas = 100000
columnas = 100000
bloque_filas = 1000  # Escribimos de 1,000 en 1,000 para ir más rápido

with open("matriz.bin", "wb") as f:
    for i in range(0, filas, bloque_filas):
        bloque = np.random.randint(0, 2, size=(bloque_filas, columnas), dtype=np.uint8)
        f.write(bloque.tobytes())
```
